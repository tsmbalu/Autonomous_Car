```python
from detecto import core, utils, visualize
#Custom dataset path
dataset = core.Dataset('LabelImage/object_detection/')
#Custom labels list marked in the dataset to be trained 
model = core.Model(['traffic_signal', 'sign_board'])
model.fit(dataset)
```

```python
image = utils.read_image('image1.jpg')
predictions = model.predict(image)
labels, boxes, scores = predictions
# ['traffic_signal', 'sign_board']
print(labels) 
print(boxes)
print(scores)
visualize.show_labeled_image(image, boxes, labels)
```

```python
model.save('traffic_signal_model.pth')
```
